TASK 1

```python
## Accept two numbers from the user
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
```

```python
# Arithmetic operations
print("\n--- Arithmetic Operations ---")
print(f"Addition: {num1} + {num2} = {num1 + num2}")
print(f"Subtraction: {num1} - {num2} = {num1 - num2}")
print(f"Multiplication: {num1} * {num2} = {num1 * num2}")
print(f"Division: {num1} / {num2} = {num1 / num2}")
print(f"Modulus: {num1} % {num2} = {num1 % num2}")
print(f"Power: {num1} ** {num2} = {num1 ** num2}")
```

```python
# Comparison and logical operators
print("\n--- Comparison & Logical Operators ---")
print(f"{num1} > {num2} : {num1 > num2}")
print(f"{num1} < {num2} : {num1 < num2}")
print(f"{num1} == {num2} : {num1 == num2}")
print(f"{num1} != {num2} : {num1 != num2}")
print(f"({num1} > 0 and {num2} > 0): {num1 > 0 and num2 > 0}")
print(f"({num1} > 0 or {num2} > 0): {num1 > 0 or num2 > 0}")
print(f"not({num1} > {num2}): {not (num1 > num2)}")
```

TASK 2

```python
import random
import numpy as np
```

```python
# Create list of 10 random integers
numbers = [random.randint(1, 100) for _ in range(10)]
print("\nOriginal List:", numbers)
```

```python
## Add and remove elements
numbers.append(50)
numbers.pop()
print("Updated List:", numbers)
## Max, Min and Sorting
print("Maximum:", max(numbers))
print("Minimum:", min(numbers))
numbers.sort()
print("Sorted List:", numbers)
```

```python
## Convert to NumPy array and calculate statistics
arr = np.array(numbers)
print("\n--- NumPy Calculations ---")
print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard Deviation:", np.std(arr))
```

TASK 3

```python
# Student dictionary
students = {
    "name": ["Meghali", "Abhrajit", "Arghyajit", "Soumendu", "Sudip", "Gobey"],
    "course": ["AI & ML", "CSE", "Data Science", "AI & ML", "Data Science", "CSE"],
    "marks": [82, 91, 67, 74, 55, 60]
}
students['grades'] = []

for marks in students["marks"]:
    if marks > 90:
        students['grades'].append("A+")
    elif marks > 80:
        students['grades'].append("A")
    elif marks > 70:
        students['grades'].append("B")
    elif marks > 60:
        students['grades'].append("C")
    else:
        students['grades'].append("D")
```

```python
for i in range(len(students["name"])):
    print(
        f"Name: {students['name'][i]}, "
        f"Course: {students['course'][i]}, "
        f"Marks: {students['marks'][i]}, "
        f"Grade: {students['grades'][i]}"
    )
```

```python
ai_tools_1 = {"ChatGPT", "TensorFlow", "PyTorch", "Scikit-learn"}
ai_tools_2 = {"ChatGPT", "Keras", "OpenCV", "PyTorch"}
print("\nAI Tools set 1:", ai_tools_1)
print("AI Tools set 2:", ai_tools_2)
```

TASK 4

```python
# Writing student details to file
with open("ai_students.txt", "w") as file:
    file.write("Meghali,78,B\n")
    file.write("Abhrajit,88,A\n")
    file.write("Arghyajit,65,C\n")
    file.write("Soumendu,92,A\n")
    file.write("Sudip,74,B\n")
    file.write("Gobey,86,B\n")
```

```python
# Reading file and displaying students with marks > 75
print("\n--- Students Scoring Above 75 ---")
with open("ai_students.txt", "r") as file:
    for line in file:
        name, marks, grade = line.strip().split(",")
        if int(marks) > 75:
            print(f"{name} - Marks: {marks}, Grade: {grade}")
```
